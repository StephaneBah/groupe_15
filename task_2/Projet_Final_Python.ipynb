```python
#Importation des bibliothèques nécessaires
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importation du dataset
data = 'Housing.csv'
data = pd.read_csv(data)
```

```python
#Création du graphique (histogramme)
plt.hist(data.bedrooms)
plt.xlabel('bedrooms')
plt.ylabel('nombres')
plt.title('boite')
plt.show()
```

## A partir de la visualisation de l'histogramme, nous constatons que la majorité des maisons de notre dataset sont à trois chambres, nous avons aussi en moyenne des maisons à deux chambres et une minorité de maisons à quatre chambres. Le reste des maisons à cinq, six chambres sont en faible minorité.

```python
#Création du graphique (nuage de points)
plt.scatter(data.area, data.price)
plt.xlabel('area')
plt.ylabel('price')
plt.title('boite')
plt.show()
```

## A partir de la visualisation du nuage de point, on constate que les prix des maisons ne dépendent pas de leurs superficies car dans notre dataset des maisons ont mêmes superficies mais pas le même prix et aussi on remarque des points un peu isolé du groupement de points, ce qui nous laisse penser que d'autres facteurs que la surface peuvent également influencer le prix d'une maison, tels que l'emplacement, l'état de la maison, les équipements, etc.
